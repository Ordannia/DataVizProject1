```python
import os
import pandas as pd
import csv
import matplotlib.pyplot as plt
import numpy as np

# Specify the full path to your CSV file
csv_path = os.path.join("..", "CountyandAgeGroupuptodate", "county_and_age_group_uptodate.csv")

# Read the CSV file into a DataFrame
age_groups_12_18_65 = pd.read_csv(csv_path, encoding="utf-8")
age_groups_12_18_65_clean = age_groups_12_18_65.dropna()

# Optionally, save the cleaned DataFrame to a new CSV file
age_groups_12_18_65_clean.to_csv('age_groups_12_18_65_clean.csv', index=False)
age_groups_12_18_65_clean.head(5)
```

```python

plt.figure(figsize=(20, 6))

# PLOT
plt.plot(age_groups_12_18_65_clean['County'], age_groups_12_18_65_clean['Percent of 12+ Pop Fully Vaccinated - Resident'], label='12+', marker='o', linestyle='-', color='b')
plt.plot(age_groups_12_18_65_clean['County'], age_groups_12_18_65_clean['Percent of 18+ Pop Fully Vaccinated - Resident'], label='18+', marker='s', linestyle='--', color='g')
plt.plot(age_groups_12_18_65_clean['County'], age_groups_12_18_65_clean['Percent of 65+ Pop Fully Vaccinated - Resident'], label='65+', marker='^', linestyle='-.', color='r')


# FORMAT

plt.xlabel('Colorado Counties', fontsize=12)
plt.ylabel('Percent Fully Vaccinated', fontsize=12)
plt.title('Percent Fully Vaccinated Per Age Groups Per County (1/9/21-5/8/23)-Line', fontsize=14)
plt.legend(title='Age Group', fontsize=10)

plt.grid(True)

plt.yticks(fontsize=10)

plt.xticks(fontsize=10)
plt.xticks(rotation=80)
plt.xticks(np.arange(0, 66, step=1))

#SHOW
plt.show()

```

```python
# Bar width
bar_width = 0.2

# Positioning of the bars on the x-axis
r1 = np.arange(len(age_groups_12_18_65_clean['County']))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

plt.figure(figsize=(20, 6))

# Create the bar chart
plt.bar(r1, age_groups_12_18_65_clean['Percent of 12+ Pop Fully Vaccinated - Resident'], color='b', width=bar_width, edgecolor='grey', label='12+')
plt.bar(r2, age_groups_12_18_65_clean['Percent of 18+ Pop Fully Vaccinated - Resident'], color='g', width=bar_width, edgecolor='grey', label='18+')
plt.bar(r3, age_groups_12_18_65_clean['Percent of 65+ Pop Fully Vaccinated - Resident'], color='r', width=bar_width, edgecolor='grey', label='65+')

# FORMAT

plt.xlabel('Colorado Counties', fontweight='bold')
plt.xticks(range(len(age_groups_12_18_65_clean['County'])), age_groups_12_18_65_clean['County'])
plt.xticks(np.arange(0, 66, step=1))
plt.xticks(rotation=80)


plt.ylabel('Percent Fully Vaccinated')
plt.title('Percent Fully Vaccinated Per Age Groups Per County (1/9/21-5/8/23)-Bar')

plt.legend(title='Age Group', fontsize=10)

# SHOW
plt.show()
```
